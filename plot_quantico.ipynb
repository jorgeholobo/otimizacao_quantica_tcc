```python
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit.circuit.library import EfficientSU2
from qiskit_algorithms.optimizers import COBYLA
from qiskit_algorithms import SamplingVQE

from IPython.display import clear_output

import plotly.graph_objects as go
import plotly.express as px

import pandas as pd
import pickle
```

```python
resultados_vqe_4 = pd.DataFrame()

with open('pickles/resultados_vqe_4_diferentes_metodos_parcial_s.pickle', 'rb') as file:
    resultados_vqe_4 = pickle.load(file)

programa_quadratico_4 = None

with open('pickles/programa_quadratico_4.pickle', 'rb') as file:
    programa_quadratico_4 = pickle.load(file)
```

```python
resultados_vqe_4.head()
```

```python
parametros_iniciais = resultados_vqe_4.loc[resultados_vqe_4.index == ('COBYLA','EfficientSU2'), 'parametros'].to_list()[0]
otimizador = COBYLA(maxiter=100)
ansatz = EfficientSU2(4)
seed = 123
```

## local

```python
sampler_local = Sampler()
```

```python
lista_energia_aux_local = []
lista_parametros_aux_local = []

def guardar_parametros_chamada_custo(eval_count, parameters, mean, std):
    lista_energia_aux_local.append(mean)
    lista_parametros_aux_local.append(parameters)
    clear_output(wait=False)

vqe_local = SamplingVQE(sampler=sampler_local, ansatz=ansatz, optimizer=otimizador, initial_point=parametros_iniciais, callback=guardar_parametros_chamada_custo)
vqe_local.random_seed = seed
vqe_eigen_local = MinimumEigenOptimizer(vqe_local)
resultado_local = vqe_eigen_local.solve(programa_quadratico_4)
```

```python
print(lista_energia_aux_local)
```

## simulador

```python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit_ibm_runtime import Sampler

service = QiskitRuntimeService(channel="ibm_quantum")

ambiente_backend_simulador ="ibmq_qasm_simulator"

backend_simulador = service.backend(ambiente_backend_simulador)

sampler_simulador = Sampler(backend=backend_simulador)
```

```python
lista_energia_aux_simulador = []
lista_parametros_aux_simulador = []

def guardar_parametros_chamada_custo(eval_count, parameters, mean, std):
    lista_energia_aux_simulador.append(mean)
    lista_parametros_aux_simulador.append(parameters)
    clear_output(wait=False)

vqe_simulador = SamplingVQE(sampler=sampler_simulador, ansatz=ansatz, optimizer=otimizador, initial_point=parametros_iniciais, callback=guardar_parametros_chamada_custo)
vqe_simulador.random_seed = seed
vqe_eigen_simulador = MinimumEigenOptimizer(vqe_simulador)
resultado_simulador = vqe_eigen_simulador.solve(programa_quadratico_4)
```

# Plot

```python
import plotly.graph_objects as go

fig = go.Figure()

def prepare_data(lista_energia_aux, device_name):
    x_data = list(range(len(lista_energia_aux)))
    y_data = [i.real if isinstance(i, complex) else i for i in lista_energia_aux]

    print(y_data)

    fig.add_trace(go.Scatter(
        x=x_data,
        y=y_data,
        mode='lines',
        name=device_name
    ))

# Replace lista_energia_aux_quantico, lista_energia_aux_simulador, lista_energia_aux_local with actual data
prepare_data(lista_energia_aux_quantico, 'ibm_brisbane')
prepare_data(lista_energia_aux_simulador, 'ibmq_qasm_simulator')
prepare_data(lista_energia_aux_local, 'local')

fig.update_layout(
    xaxis_title='Chamada de funo de custo',
    yaxis_title='Energia',
    legend_title='Computadores',
    title='',
    width=1400,
    height=700, 
    font=dict(size=35), 
    margin=dict(l=90, r=90, t=90, b=90)
)

nome_figura = "figuras_dados/comparacao_quantico.png"

fig.write_image(nome_figura, width=1400, height=700)
fig.show()
```
